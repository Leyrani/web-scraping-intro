```python
# dependencies
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# check that the site is pingable
url = "https://webpages.charlotte.edu/mscipion/"

r= requests.get(url)
print(r)
```

```python
# set up beautiful soup
soup = BeautifulSoup(r.text, "lxml")
```

```python
# find the table, verify that i found the right thing
table = soup.find("table")
print(table)
```

```python
# find column headers
headers = table.find_all("th")
print(headers)
```

```python
# turn the column headers into a list
titles = []

for i in headers:
    title = i.text
    titles.append(title)
    
print(titles)
```

```python
# turn the table into a DataFrame
scraped_table= pd.DataFrame(columns=titles)
print(scraped_table)
```

```python
# find all of the rows of the table
rows = table.find_all("tr")
print(rows)
```

```python
# print all of the rows
for i in rows[1:]:
    print(i)
```

```python
# print the rows again, but make them look more like rows
for i in rows[1:]:
    data = table.find_all("td")
    print(data)
```

```python
# print the rows again, but make them look even more like rows
for i in rows[1:]:
    data = table.find_all("td")
    row = [tr.text for tr in data]
    print(row)
```

```python
# turn everything we've done into a dataframe with the scraped table
# using list comprehension to identify the rows and add the data to the dataframe created earlier
for i in rows[1:]:
    data = i.find_all("td")
    row = [tr.text for tr in data]
    l =len(scraped_table)
    scraped_table.loc[l]= row

print(scraped_table)
```
